##Tokenize function

```python
def tokenize(example):
    model_inputs = tokenizer(
        example["input"],
        padding="max_length",
        truncation=True,
        max_length=128
    )
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(
            example["output"],
            padding="max_length",
            truncation=True,
            max_length=128
        )
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
```

##Tokenize dataset



```python
from datasets import load_dataset

# Load dataset
dataset = load_dataset("salahyahya/grammeranderrordataset")

# Tokenize
tokenized_dataset = dataset.map(tokenize, batched=True)
```

##Split into train and eval

```python
split_dataset = tokenized_dataset["train"].train_test_split(test_size=0.1)
train_dataset = split_dataset["train"]
eval_dataset = split_dataset["test"]
```

## Training Arguments

```python
from transformers import Seq2SeqTrainingArguments

training_args = Seq2SeqTrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    weight_decay=0.01,
    save_total_limit=3,
    num_train_epochs=3,
    predict_with_generate=True,
    logging_dir="./logs",
    logging_steps=10,
)
```

##Trainer


```python
from transformers import Trainer

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    tokenizer=tokenizer,
)
```

# Train the model

```python
trainer.train()
```

## result accuracy

```python
results = trainer.evaluate()
print(results)
```

## Gradio deployment

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
import gradio as gr

# Load the model and tokenizer from local directory
model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-small")
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-small")

# Create the pipeline
pipe = pipeline("text2text-generation", model=model, tokenizer=tokenizer)

# Gradio UI function
def correct_grammar(text):
    output = pipe(text, max_length=128, clean_up_tokenization_spaces=True)
    return output[0]["generated_text"]

# Launch Gradio
gr.Interface(fn=correct_grammar, inputs="text", outputs="text", title="Grammar Correction LLM").launch()
```
